```python
#Import pandas,json and numpy
import pandas as pd
import json
from pandas.io.json import json_normalize
import numpy as np
```

```python
#Read file and save
df = pd.read_json('data/world_bank_projects.json')
#Inspect
df.info
```

```python
# Set options to inspect ALL of the columns:
pd.options.display.max_columns = 50

# Inspect the structure of the dataframe:
df.head()
```

```python
# Save top ten countries
top_ten_countries = df.countryshortname.value_counts().head(10)
top_ten_countries
```

```python
#  loadingjson and normalizing

df = json.load((open('data/world_bank_projects.json')))

df = json_normalize(df, 'mjtheme_namecode')
df.head(15)
```

```python
# Save the total count of each project code:
project_count_per_theme = df.code.value_counts()

# Save only the top ten and inspect:
top_ten_theme_count = project_count_per_theme.head(10)
top_ten_theme_count
```

```python
# Theme name value count; Some Names are empty
#  loading in the raw json to wrangle the nested fields later
with open('data/world_bank_projects.json') as json_file:
    json_data = json.load(json_file)
df_themes = json_normalize(json_data,'mjtheme_namecode', ['id'])
print(df_themes.name.value_counts().head(10))
```

```python
# Sort by code, then name and inspect:
df = df.sort_values(['code', 'name'])
df.head(10)
```

```python
# Replace empty name values with NaN and inspect:
df.name[df['name'] == ''] = np.nan
df.head(10)
```

```python
# Backfill NaN values and inspect:
df = df.fillna(method='bfill')
df.head()
```
